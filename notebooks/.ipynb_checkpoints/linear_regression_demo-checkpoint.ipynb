# Linear Regression Demo (NumPy)

This notebook demonstrates a **Linear Regression model from scratch** using NumPy.

- Load dataset (`housing.csv`)
- Preprocess features
- Train using gradient descent
- Evaluate and visualize predictions

This is a beginner-friendly, step-by-step demo.


```python
import sys
import os

sys.path.append(os.path.abspath(".."))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from src.linear_regression import LinearRegression
```

```python
data_path = "../data/housing.csv"
data = pd.read_csv(data_path)
data.head()
```

```python
print(data.info())
print(data.describe())

numeric_cols = data.select_dtypes(include=np.number).columns.tolist()
numeric_cols.remove("price")
data[numeric_cols].hist(figsize=(12,8))
plt.show()
```

```python
target_column = "price"

X = data.select_dtypes(include=np.number).drop(columns=[target_column],errors="ignore").fillna(0).values
y = data[target_column].values

X_mean = X.mean(axis=0)
X_std = X.std(axis=0)
X_norm = (X-X_mean)/X_std
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.2, random_state=42)

print("Train/Test shapes:", X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model = LinearRegression(learning_rate=0.01, n_epochs=1000)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

mse = np.mean((y_test - y_pred) ** 2)
print(f"Test MSE: {mse:.4f}")

plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', linewidth=2)
plt.xlabel("True Price")
plt.ylabel("Predicted Price")
plt.title("Predicted vs Actual")
plt.show()
```
